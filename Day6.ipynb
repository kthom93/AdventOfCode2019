```
def countOrbits(orbits):
  orbits = orbits.split('\n')
  d = {}
  s = set()
  for orbit in orbits:
    center, outer = orbit.split(')')
    if not center in d:
      d[center] = []
    d[center].append(outer)
    s.add(center)
    s.add(outer)

  total = 0
  for planet in s:
    total += countR(planet, d)
  return total

def countR(planet, d):
  if not planet in d.keys():
    return 0
  total = len(d[planet])
  for i in range(len(d[planet])):
    total += countR(d[planet][i], d)
  return total
```

```
def countTransfers(orbits):
  orbits = orbits.split('\n')
  d = {}
  s = set()
  for orbit in orbits:
    center, outer = orbit.split(')')
    if not center in d:
      d[center] = []
    d[center].append(outer)
    s.add(center)
    s.add(outer)
  shortest = 99999
  for planet in s:
    transfers = distanceTo(d, planet,'YOU') + distanceTo(d, planet, 'SAN')
    shortest = min(shortest, transfers)

  return shortest - 2

def distanceTo(d, start, target):
  if start == target:
    return 0
  if start not in d:
    return 99999
  distances = []
  for planet in d[start]:
    distances.append(distanceTo(d, planet, target))
  return min(distances) + 1
```
